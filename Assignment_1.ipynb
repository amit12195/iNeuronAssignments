3.) Describe three different data types.

1.> Numeric Data Type:

   In Python, numeric data type represents the data that has a numeric value like integer, floating number, and complex number.
ex:
a = 10    # integer
b = 2.9   # float 
c = 5+4j  # complex

2.> Sequence Data Type:

sequence is the ordered collection of similar or dissimilar data types, like String, List, Tuple,..etc
String : Any sequence of characters within either single quotes or double quotes or triple single quotes is considered as a   String.

       Syntax: s = 'Welcome to India'
       
List:  List is a group of individual objects as a single entity where insertion order preserved and duplicates are allowed.
	   heterogeneous objects are allowed.
	   List is dynamic.
	   ex: list1 = [10,20,'Delhi']
       
Tuple: Tuple is exactly same as List except that it is immutable, or we can say "Tuple is Read only version of List".
       ex: t = (10,12,14)

3.> Set Data type:

    Set is a group of unique values.
    Duplicates are not allowed.
    Insertion order is not preserved.But we can sort the elements.
    Indexing and slicing not allowed for the set.
    Heterogeneous elements are allowed.
    Set objects are mutable .
    
    ex: s = {10,20,30,40,50}
