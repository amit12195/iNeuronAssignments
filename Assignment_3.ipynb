2.) When does the code in a function run: when its specified or when it's called?

Answer:


Whenever we call a function then program leaves the current section of code and a new stack frame is created with all the function’s data and this stack frame is pushed in the program stack, and the stack pointer that always points the top of the program stack points the stack frame pushed as it is on the top of the program stack.

Internally function’s execution follow below steps:

a.) Stack Frame is pushed into stack.
b.) Sub-routine instructions are executed.
c.) Stack Frame is popped from the stack.
d.) Now Program Counter is holding the return address.
